```python
import pandas as pd
data = pd.read_csv("Student-University.csv")
data
```

```python
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()






shuffle = data.sample(frac=1)
t_size = int(0.7 * len(data))
t_set = shuffle[:t_size]
tt_set = shuffle[t_size:]
tt_set
```

```python
from random import random
import numpy as np
x1 = t_set["X1"].values
x2 = t_set["X2"].values
y_train = t_set["y"].values

b0 = 0.0
b1 = 0.0
b3= 0.0
pred = 0
y=[]
print(y_train)
for i in range(50000):
    for i in range(len(y_train)):
        y_hat = b0+(b1*x1[i]) +(b3 *x2[i])
        pred=1/(1+np.exp(-y_hat))
        b0 = b0+0.01*(y_train[i]-pred)*pred*(1-pred)*1
        b1 = b1+0.01*(y_train[i]-pred)*pred*(1-pred)*x1[i]
        b3 = b3+0.01*(y_train[i]-pred)*pred*(1-pred)*x2[i]
    #print("COF",b0,b1,b3)
#print(pred)
```

```python
xt1 = tt_set["X1"].values
xt2 = tt_set["X2"].values
y_test = tt_set["y"].values
y=[]
print(y_test)
    
for i in range(len(y_test)):
    y_hat = b0+(b1*xt1[i]) +(b3 *xt2[i])
    pred=1/(1+np.exp(-y_hat))
    pred = int(pred)
    y.append(pred)
print(y)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y))
```
